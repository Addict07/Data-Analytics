```python
all_data = pd.read_csv("/content/all_data.csv")
```

**DATA CLEANING!**

```python
all_data.isnull().sum()/all_data.shape[0]
```

```python
#checking for null rows in the datafram
n_df = all_data[all_data.isna().any(axis = 1)]
n_df
```

```python
all_data.isnull().sum()
```

```python
#now we drop rows only if all the rows have null values

all_data = all_data.dropna(how = 'all')
all_data.head()
```

**WE CREATE A NEW COLUMN FOR MONTHS**

```python
all_data.columns
```

```python
all_data['Months'] = all_data['Order Date'].str[0:2]
```

```python
temp_df = all_data[all_data['Order Date'].str[0:2]== 'Or']
temp_df
```

```python
#We then use a smart way to drop the places with 'Or'
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
```

```python
all_data[all_data.isna().any(axis = 1)]
```

```python
all_data.isnull().sum()
```

```python
all_data = all_data.dropna(how = 'any')
```

```python
all_data[all_data.isna().any(axis = 1)]
```

```python
all_data['Months'] = all_data['Months'].astype('int32')
```

```python
all_data.dtypes
```

```python
#Now we convert the months to an integer datatype
all_data.isnull().sum()
```

```python
#We convert the columns to their specific types before we can multiply
all_data['Quantity Ordered'] = all_data['Quantity Ordered'].astype('int32')
all_data['Price Each'] = all_data['Price Each'].astype('float64')
```

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
```

```python
all_data
```

** Q1 The best month for sales and how much was earned?**

```python
#The groupby function helps in this exercise

monthly_sales = all_data.groupby('Months').sum()
```

**WE CAN SEE BELOW THAT THE BEST MONTH FOR SALES IS DECEMBER**

```python
monthly_sales
```

```python
import matplotlib.pyplot as plt
months = range(1,13)

plt.bar(months,monthly_sales['Sales'])
plt.title("A plot of Sales vs MOnths")
plt.xlabel("Months")
plt.ylabel("Monthly  Sales in USD ($)")
plt.show()
```

**Q2 WHICH CITY HAD THE HIGHES AMOUNT OF SALES**

```python
#WE THEN SPLIT ACCESS THE SECOND INDEX BASED ON THE SPLIT OF ADDRESS
#USING THE APPLY()
#APPLY(LAMBDA X: X.SPLIT(,))
all_data['Cities'] = all_data['Purchase Address'].apply(lambda x: x.split(',')[1])
```

```python
all_data
```

```python
sales_by_city = all_data.groupby('Cities').sum()
```

```python
sales_by_city
```

```python
city_op = [city for city , df in all_data.groupby('Cities')]
plt.bar(city_op,sales_by_city['Sales'])
plt.xticks(city_op, rotation = 'vertical', size = 8)
plt.xlabel("Cities")
plt.ylabel("Sales in USD ($)")
plt.show()
```

```python
df = all_data[all_data['Order ID'].duplicated(keep = False)]
```

```python
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df.head(20)
```

```python
df = df[['Order ID', 'Grouped']].drop_duplicates()
```

```python
df
```

```python
from itertools import combinations
from collections import Counter
```

```python
count = Counter()
```

```python
for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))
count.most_common(10) 
```

```python
for key, value in count.most_common(10):
    print(key,value)
```

```python
product_group = all_data.groupby('Product')
quantity_ordered = product_group.sum()['Quantity Ordered']

keys = [pair for pair, df in product_group]
plt.bar(keys, quantity_ordered)
plt.xticks(keys, rotation='vertical', size=8)
plt.show()
```

```python
prices = all_data.groupby('Product').mean()['Price Each']

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(keys, quantity_ordered, color='r')
ax2.plot(keys, prices, color='b')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(keys, rotation='vertical', size=8)
```
